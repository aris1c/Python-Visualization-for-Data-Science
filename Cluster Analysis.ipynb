# Cluster Analysis with python : A simple Example

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

buat table mainan, n sample buat 300 baris dan 2 kolom

```python
X, y_true = make_blobs(n_samples=300,
                       n_features=2,
                       centers=4,
                       cluster_std=0.7,
                       random_state=0)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
Kmeans = KMeans(n_clusters=4)

Kmeans.fit(X) #menyuruh kmeans belajar dari si X
```

```python
y_kmeans = Kmeans.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],s=50,c=y_kmeans) #c=y_kmeans itu memisahkan membuat model membagi menjadi 5 kelompok dan otomatis berubah warna
centers = Kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black')
```
